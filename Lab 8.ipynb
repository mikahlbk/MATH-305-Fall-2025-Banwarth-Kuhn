$\textbf{Exercise 2.}$ The code below sorts a list of numbers from smallest to largest. 

$\textbf{First}$, trace through the for loop to understand how it works! This is good practice for everyone and especially those of you that are still working on Learning Goal 4 :)

$\textbf{Second}$, edit the code so it uses a function to swap elements.

```python
# list of nums
numbers = [5, 2, 1]

# iterate through the indices of the list
for i in range(0,len(numbers)):
    # considers the current numbers in indices 0 through len(numbers) - i - 1  
    for j in range(0, len(numbers) - i - 1):
        # compare current numbers in indices j and j+1 (the numbers in each index will change if a swap is made)
        # if a larger number is to the left swap the two numbers
        if numbers[j] > numbers[j + 1]:
            # swap the elements
            temp = numbers[j]
            numbers[j] = numbers[j + 1]
            numbers[j + 1] = temp

print("Sorted list:", numbers)
```
