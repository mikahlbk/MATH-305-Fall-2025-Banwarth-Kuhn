Consider the function
$$f(x) = x^{3}-6x^{2}+11x-6.$$

Run the next code cell to plot $f(x)$.

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x:x**3-6*x**2+11*x-6
x_vals = np.linspace(0,4,500)
y_vals = f(x_vals)

plt.plot(x_vals,y_vals,color = "black",label = "f(x)")
plt.grid(True)
```
