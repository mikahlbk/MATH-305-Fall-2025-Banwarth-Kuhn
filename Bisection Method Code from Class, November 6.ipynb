```python
#intial check cell
import numpy as np
error_vec = []
#Inputs: f lambda function, a, b, number iterations, epsilon/threshold
#Outputs: the approximate root
def my_bisection_method(f,a,b,num_iters,epsilon):
    if f(a)*f(b) > 0:
        print("IVT not met")
        return None
    if np.abs(f(a))<epsilon:
        print("Endpoint {} is the root".format(a))
        return a
    if np.abs(f(b))<epsilon:
        print("Endpoint {} is the root".format(b))
        return b
    p_curr = 0
    for i in range(1,num_iters):
        print("iteration",i)
        p_curr = a + (b-a)/2
        #print(f(p_curr))
        if np.abs(f(p_curr))<epsilon:
            print("Root found, p = {}".format(p_curr))
            return p_curr
        if f(p_curr)*f(a) < 0:
            b = p_curr
        else:
            a = p_curr
        error_vec.append(np.sqrt(2)-p_curr)
        print("p_curr =",p_curr)

f = lambda x:x**2-2
a = 0
b = 2
epsilon = 1e-3#scientific notation
num_iters = 100
root = my_bisection_method(f,a,b,num_iters,epsilon)
print("The root is ",root)
```
