Consider the function
$$f(x) = \cos(x)-x$$

Run the next code cell to plot $f(x)$.

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: np.cos(x) - x
x_vals = np.linspace(-.25, 1, 500)
y_vals = f(x_vals)

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, color="black", label="f(x)")
ax.grid(True)

# --- Move the x-axis to y=0 ---
ax.spines['bottom'].set_position('zero')

# Optional: Move y-axis to x=0
ax.spines['left'].set_position('zero')

# Add labels/legend
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.legend()

plt.show()
```
